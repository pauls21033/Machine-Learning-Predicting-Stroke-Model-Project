```python
import pandas as pd
import sqlalchemy
db = sqlalchemy.create_engine('sqlite:///stroke_analysis.sqlite3')
```

```python
df = pd.read_csv('Resources/Stroke_analysis Cleaned2.csv')
```

```python
df
```

```python
stroke_df = df.drop(columns=["severity_desc"])
```

```python
gender_dict = {'Male': 0, 'Female': 1}
stroke_df = stroke_df.replace({'gender': gender_dict})
```

```python
stroke_df
```

```python
stroke_df.to_sql('stroke_analysis', db, if_exists="replace")
```
